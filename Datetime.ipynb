```python
import pandas as pd
```

```python
df = pd.read_csv('./data/temperatures.csv')
df.head()
```

```python
df.info()
```

```python
# Datefmf datetime으로 바꾸기
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```python
df.info()
```

## 1. 특정 요일에 해당하는 데이터 조회하기

```python
## 월요일 데이터 조회 월~일 = 0~6
##또는 df[df['Date'].dt.dayofweek==0]
df.query('Date.dt.dayofweek==0')
```

```python
#월,수,금
##또는 df[df['Date'].dt.dayofweek.isin.(datofweek)]
dayofweek = [0, 2, 4]
df.query('Date.dt.dayofweek in @dayofweek')
```

## 2. 특정 날짜에 해당하는 데이터 조회하기
- 특정 날짜

```python
target_date = '1981-01-02'
df.query('Date == @target_date')
## 또는 df[df['Date']==target_date]
```

### 특정 연도, 월, 일

```python
target_year = 1990
df.query('Date.dt.year == @target_year')
## 또는 df[df['Date'].dt.year==target_year]
```

```python
target_month = 9
df.query('Date.dt.month == @target_month')

##df[df['Date'].dt.month==target_day]
```

```python
target_day=13
df.query('Date.dt.day==@target_day')

## df[df['Date'].dt.day==target_day]
```

## 3. 특정 날짜 범위에 해당하는 데이터 조회

```python
## 1981년 1월 1일부터 1981년 1월 4일까지 데이터 조회

df.query('"1981-01-01"<= Date <= "1981-01-04"')
## 또는 df[(df['Date']>='1981-01-01') & (df['Date']<='1981-01-04')]
```

```python
## 1987년 부터 1989년까지의 데이터 조회

df.query('1987 <= Date.dt.year <= 1989') 
## 또는 df[(df['Date'].dt.year >= 1987) & (df['Date'].dt.year <= 1989)]
```

## 그룹화 하기

```python
## 연도별로 그룹화하기
df['Year'] = df['Date'].dt.year #Year 칼럼 만들기
df.groupby('Year')['Temp'].aggregate(['max','min','mean']).reset_index()
```

```python
## 요일별로 그룹화하기
num_to_day = {
    0 : '월요일',
    1 : '화요일',
    2 : '수요일',
    3 : '목요일',
    4 : '금요일',
    5 : '토요일',
    6 : '일요일',
}
 
df['Day'] = [num_to_day[k] for k in df['Date'].dt.dayofweek] ## 요일 칼럼
df.groupby('Day')['Temp'].aggregate(['max','min','mean']).reset_index()
```
